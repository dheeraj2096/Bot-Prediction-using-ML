```python
#importing packages to apply classifiers
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
import seaborn as sns
```

```python
gnb = GaussianNB()
tree = DecisionTreeClassifier(max_depth=2)
```

```python
used_features = ['Total_tweets','Tweets_per_Day']
```

```python
#training the models
gnb.fit(train_data[used_features],train_data['Bot'])
tree.fit(train_data[used_features],train_data['Bot'])
```

```python
#predicting the values
pred_gnb = gnb.predict(test_data[used_features])
pred_tree = tree.predict(test_data[used_features])
```

```python
expected = test_data['Bot']
```

```python
test_index = [i for i in range(0,75)]
```

```python
#plotting the observed values vs Predictions
plt.plot(test_index,pred_gnb,label="Predicted Values")
plt.plot(test_index,expected,label="Expected Values")
plt.xlabel("Index Values")
plt.ylabel("Tweet_Count_per_Day")
plt.title("Using Gaussian Naive Bayes Classifier")
plt.legend()
```

```python
#plotting the observed values vs Predictions
plt.plot(test_index,pred_tree,label="Predicted Values")
plt.plot(test_index,expected,label="Expected Values")
plt.xlabel("Index Values")
plt.ylabel("Tweet_Count_per_Day")
plt.title("Using Decision Tree Classifier")
plt.legend()
```

```python
#Calculating the MSE for GNB
gnb_error = metrics.mean_squared_error(pred_gnb,expected)
print(gnb_error)
```
